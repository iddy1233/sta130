````python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate Sample Data
np.random.seed(42)  # Ensures random numbers
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Create a sample of 100 data points with mean=50 and SD=10

# 2. Define the Bootstrap Function
def bootstrap_confidence_interval(data, statistic, num_iterations=10000, confidence_level=0.95):
    """
    Calculate the bootstrap confidence interval for a specified statistic.
    
    Parameters:
    - data (array-like): The original sample data.
    - statistic (function): The statistic to compute (e.g., np.mean, np.median).
    - num_iterations (int): Number of bootstrap samples to generate.
    - confidence_level (float): Desired confidence level (e.g., 0.95 for 95%).
    
    Returns:
    - tuple: (lower_bound, upper_bound) of the confidence interval.
    """
    boot_stats = []  # List to store the computed statistics from each bootstrap sample
    n = len(data)    # Original sample size
    
    for _ in range(num_iterations):
        # Resample the data with replacement
        boot_sample = np.random.choice(data, size=n, replace=True)
        # Calculate the desired statistic for the bootstrap sample
        stat = statistic(boot_sample)
        # Append the result to the list of bootstrap statistics
        boot_stats.append(stat)
    
    # Convert the list of bootstrap statistics to a NumPy array for percentile calculations
    boot_stats = np.array(boot_stats)
    
    # Calculate the lower and upper percentiles for the confidence interval
    lower_percentile = ((1.0 - confidence_level) / 2.0) * 100
    upper_percentile = (confidence_level + (1.0 - confidence_level) / 2.0) * 100
    
    # Determine the confidence interval bounds
    lower_bound = np.percentile(boot_stats, lower_percentile)
    upper_bound = np.percentile(boot_stats, upper_percentile)
    
    return lower_bound, upper_bound

# 3. Compute 95% Confidence Interval for the Mean
mean_ci = bootstrap_confidence_interval(sample_data, np.mean)
print(f"95% Bootstrap Confidence Interval for the Mean: {mean_ci}")

# 4. Compute 95% Confidence Interval for the Median
median_ci = bootstrap_confidence_interval(sample_data, np.median)
print(f"95% Bootstrap Confidence Interval for the Median: {median_ci}")

# 5. (Optional) Visualize Bootstrap Distribution for the Mean
def plot_bootstrap_distribution(data, statistic, ci, num_iterations=10000):
    """
    Plot the bootstrap distribution of a statistic with confidence interval.
    Parameters:
    - data (array-like): The original sample data.
    - statistic (function): The statistic to compute (e.g., np.mean, np.median).
    - ci (tuple): Confidence interval bounds.
    - num_iterations (int): Number of bootstrap samples.
    """
    boot_stats = [statistic(np.random.choice(data, size=len(data), replace=True)) 
                 for _ in range(num_iterations)]
    
    plt.hist(boot_stats, bins=30, edgecolor='k', alpha=0.7)
    plt.axvline(ci[0], color='red', linestyle='--', label=f"{int((1-0.95)/2*100)}th Percentile")
    plt.axvline(ci[1], color='red', linestyle='--', label=f"{int((0.95 + (1-0.95)/2)*100)}th Percentile")
    plt.title(f'Bootstrap Distribution of the {"Mean" if statistic == np.mean else "Median"}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Mean
plot_bootstrap_distribution(sample_data, np.mean, mean_ci)
# Median
plot_bootstrap_distribution(sample_data, np.median, median_ci)
````
