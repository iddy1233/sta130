Simple Linear Regression: model the relationship between a single predictor variable Xi and an outcome variable,Yi. the relationship can be written as Yi=β0+β1Xi+ϵ. 
β0 is the interception of regression line, β1 is slope. ϵ is the error of every term's observation, introduces randomness into the model. error is assumed to follow a normal distribution with mean of 0.

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set parameters
n = 100  # number of observations
beta0 = 2.0  # intercept
beta1 = 3.0  # slope
sigma = 1.0  # standard deviation of error

# Generate predictor variable X
x = np.random.uniform(0, 10, n)

# Generate error terms (normally distributed)
errors = np.random.normal(0, sigma, n)

# Generate outcome variable Y using the simple linear regression equation
Y = beta0 + beta1 * x + errors

# Create a scatter plot to visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(x, Y, color='blue', label='Observed Data', alpha=0.6)
plt.plot(x, beta0 + beta1 * x, color='red', label='True Regression Line', linewidth=2)
plt.title('Simple Linear Regression Simulation')
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.legend()
plt.grid()
plt.show()
````

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf 
# used for statistical modeling,provides functions to fit various types of statistical models, including OLS regression, using formula notation.
import plotly.express as px

# Set parameters for the simulation
n = 100  # number of observations
beta0 = 2.0  # intercept
beta1 = 3.0  # slope
sigma = 1.0  # standard deviation of error

# Generate predictor variable X
x = np.random.uniform(0, 10, n)

# Generate error terms (normally distributed)
errors = np.random.normal(0, sigma, n)

# Generate outcome variable Y using the simple linear regression equation
Y = beta0 + beta1 * x + errors

# Combine x and Y into a pandas DataFrame
df = pd.DataFrame({'x': x, 'Y': Y})

# Import the statsmodels library for regression analysis
# smf stands for statsmodels formula API which allows us to specify models using R-style formulas

# Fit the Simple Linear Regression model
model_data_specification = smf.ols("Y ~ x", data=df) 
# pecifies OLS model, "Y~x", tells statsmodels to regress Y on x
fitted_model = model_data_specification.fit()  
#  fits the model to the data, calculating the regression coefficients, and minimize the squared residuals

# Display the model summary
print(fitted_model.summary()) 
#  provides a summary of the regression results, including coefficients and statistics
print(fitted_model.summary().tables[1]) 
# provides a table of the estimated coefficients and their statistics
print(fitted_model.params)  
# returns the estimated coefficients (intercept and slope)
print(fitted_model.params.values)  
# This returns the estimated coefficients as an array
print(fitted_model.rsquared)  
# This provides the R-squared value, indicating the proportion of variance in Y explained by x

# Add a column to the DataFrame for the legend
df['Data'] = 'Data'  # Hack to add data to legend

# Create a scatter plot with the fitted regression line
fig = px.scatter(df, x='x', y='Y', color='Data', 
                 title='Fitted Simple Linear Regression Model: Y vs. x')

# Add the fitted regression line to the figure
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                 line=dict(color='blue'), name="Fitted Line")

fig.show(renderer="png")  
````

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Set parameters for the simulation
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'x': x, 'y': Y})  # Note that 'y' is lowercase

# Fit the Ordinary Least Squares model
model_data_specification = smf.ols("y ~ x", data=df)  # Use 'y' (lowercase) here
fitted_model = model_data_specification.fit()

# Add a column to the DataFrame for the legend
df['Data'] = 'Data'  # Hack to add data to legend

# Create scatter plot with the correct column names
fig = px.scatter(df, x='x', y='y', color='Data', 
                 trendline='ols', title='Y vs. x')

# Add the fitted values as a blue line
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Add the theoretical regression line (orange dotted line)
x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=f'{beta0} + {beta1} * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],  # Use 'y' here
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)

# Add horizontal line at y-bar (mean of y)
fig.add_scatter(x=x_range, y=[df['y'].mean()] * 2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

# Show the figure as a PNG
fig.show(renderer="png")
````

The Ordinary Least Squares (OLS) method is a statistical technique used to find the best-fitting line for a set of observed data in regression analysis, and it minimize residuals.The therotical line, which is the orange linee on the diagrams, is the line we choose. We use squares in Ordinary Least Squares because we need to eliminate the impact of negativity, and prevent total residuals close to 0, and provide a true meansure of the model's overall error. ALso, when we apply square to that, we will have a greater wight to larger errors, and this aooroach help be more senstive to outliers that different from predivtion.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define long wait time limit
long_wait_limit = 71
long_wait = old_faithful['waiting'] > long_wait_limit

# Step 1: Bootstrap Sampling Distribution of Fitted Slope Coefficients
bootstrapped_slope_coefficients = []

# Perform bootstrap resampling
for _ in range(1000):  # Number of bootstrap samples
    sample_data = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    model = smf.ols('duration ~ waiting', data=sample_data).fit()
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Convert to a numpy array for further calculations
bootstrapped_slope_coefficients = np.array(bootstrapped_slope_coefficients)

# Visualize the bootstrapped sampling distribution of slope coefficients
plt.hist(bootstrapped_slope_coefficients, bins=30, edgecolor='k')
plt.title("Bootstrapped Sampling Distribution of Slope Coefficients")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.axvline(x=0, color='red', linestyle='--', label='Slope = 0')
plt.legend()
plt.show()

# Step 2: Simulate Sampling Distribution under Null Hypothesis (β1=0)
old_faithful_simulation = old_faithful[long_wait].copy()
simulated_slope_coefficients = []

# Simulate data under the null hypothesis
for _ in range(1000):  # Number of simulations
    old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation['waiting'] + stats.norm(loc=0, scale=0.37).rvs(size=long_wait.sum())
    simulated_model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slope_coefficients.append(simulated_model.params['waiting'])

# Convert to a numpy array for further calculations
simulated_slope_coefficients = np.array(simulated_slope_coefficients)

# Visualize the simulated sampling distribution of slope coefficients under the null hypothesis
plt.hist(simulated_slope_coefficients, bins=30, edgecolor='k')
plt.title("Simulated Sampling Distribution of Slope Coefficients (Null Hypothesis)")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.axvline(x=0, color='red', linestyle='--', label='Slope = 0')
plt.legend()
plt.show()

# Step 3: Calculate the 95% Bootstrapped Confidence Interval for Slope Coefficients
bootstrapped_ci = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print("95% Bootstrapped Confidence Interval for Slope Coefficients:", bootstrapped_ci)

# Check if 0 is within the confidence interval
contains_zero = (bootstrapped_ci[0] <= 0 <= bootstrapped_ci[1])
print("Does the 95% confidence interval contain 0?", contains_zero)

# Step 4: Calculate the observed slope and the simulated p-value
observed_slope = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params['waiting']
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print("Simulated p-value for the null hypothesis (no linear association):", simulated_p_value)


````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_

> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_

> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_

````python
# a)
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf

# Load dataset and add 'kind' variable for short/long wait times
old_faithful = sns.load_dataset('geyser')
old_faithful['kind'] = ['long' if x >= 68 else 'short' for x in old_faithful['waiting']]

# Observed mean difference between 'long' and 'short'
observed_mean_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test
num_permutations = 1000
permutation_diffs = []

for _ in range(num_permutations):
    # Shuffle labels
    shuffled_labels = old_faithful['kind'].sample(frac=1, replace=False).values
    permuted_data = old_faithful.assign(kind_shuffled=shuffled_labels)
    
    # Calculate mean difference for permuted data
    perm_mean_diff = permuted_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(perm_mean_diff)

# Calculate p-value
permutation_diffs = np.array(permutation_diffs)
p_value = (np.abs(permutation_diffs) >= np.abs(observed_mean_diff)).mean()
print("Observed Mean Difference:", observed_mean_diff)
print("Permutation Test p-value:", p_value)
````

````python
# Bootstrap confidence interval
num_bootstraps = 1000
bootstrap_diffs = []

for _ in range(num_bootstraps):
    # Resample within each group
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    
    # Calculate mean difference for bootstrapped data
    boot_mean_diff = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrap_diffs.append(boot_mean_diff)

# Calculate 95% bootstrap confidence interval
bootstrap_diffs = np.array(bootstrap_diffs)
ci_lower, ci_upper = np.quantile(bootstrap_diffs, [0.025, 0.975])
print("Bootstrap 95% Confidence Interval:", (ci_lower, ci_upper))
````

In the first method, we rerange "short" and "long" variables, and we simulate the situation under the no difference hypothesis situation, and generate a possible distribution of mean differences.
Finally, the p-value help to measure whether the observed mean difference is in the extreme part of the rearranged distribution. If we have a small p-value, it suggests that the observed difference is unlikely to be caused by random chance, and we can reject the null hypothesis.

In the bootstrap method, we repeatedly resample within each group to create new samples, and then calculate the mean difference between the two groups for each new sample. We then have a distribution of the mean differences, which we can use to estimate a confidence interval. If there is no zero within the confidence interval, then there is a difference between the two groups.

Both methods estimate the mean difference in eruption duration between two groups of waiting times. However, the first method is assuming there are no difference between two froups, and try to reject the hypothesis, and the second method give a interval, and we get a range of mean differences
